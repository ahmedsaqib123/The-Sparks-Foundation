### Exploring & Observing Cites - States - Region

```python
df['State'].value_counts()
```

```python
df['City'].value_counts()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

##### Comparitivly California has the highest number of buyers and Wyoming has the lowest

```python
plt.figure(figsize=(10,8))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

##### We can observe that West has the most sale/ buyers

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

##### Max Profit ranges between 7k-8k 
##### There are mostly non discounted but after that 20% is given 
##### Mostly 2 to 3 items are sold


```python
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

##### This above plot shows the relationship between sales and profit. 

```python
#Discount and Profit Relation
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','orange'],figsize=(10,6))
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
#States wise profit loss
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','red'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['pink','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

##### Technology has the highest profit and sale
