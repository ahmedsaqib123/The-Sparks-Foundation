### Task 4: Prediction using Decision Tree Algorithm
##### Create the Decision Tree classifier and visualize it graphically.
##### The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
#Importing Libraries 
import pandas as pd 
```

```python
df = pd.read_csv('Datasets/Iris.csv')
df
```

```python
#Features here are the Species column and we want to classify it accordingly
df = df.drop(['Id'],axis=1)
df
```

```python
X = df.iloc[:, :4]
y = df.loc[:,['Species']]
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)   #divide data into test and train

model = DecisionTreeClassifier()   #call the decision tree classifier
model.fit(X_train,y_train)         #fit the data into the model
```

```python
pred = model.predict(X_test)     #predict the values using test data
accuracy_score(pred,y_test)    #calculate the accuracy by comparing the test and predicted values
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.figure(figsize=(15,12))
plot_tree(model, filled=True, rounded=True)   #plot the decision tree
```
